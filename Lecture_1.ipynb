# Importing Data with Pandas

### Objectives
- To read a csv file
- To get familiar with the data types
- To check for the missing values and displaying statistics of a dataset

### Importing Required Libraries

```python
import pandas as pd
```

```python
# Reading a csv file
df = pd.read_csv("bowler_most_wickets_ODI.csv")
```

```python
# Displaying the first five columns
display(df.head())
```

```python
# Displaying the last five columns
display(df.tail())
```

```python
# Printing the shape of the dataset
print(df.shape)
```

```python
# Printing the number of rows and columns
print("Number of rows = ", df.shape[0])

print("Number of columns = ", df.shape[1])
```

```python
# Checking for missing values and data types of each column
print(df.info())
```

```python
# Checking data statistics
display(df.describe())
```

```python
# Checking the statistics for string type columns
display(df.describe(include=['O']))
```

```python
# Printing the column names
print(df.columns)
```
